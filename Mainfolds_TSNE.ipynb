Mainfolds: T-SNE

```python
#Cargamos los paquetes necesarios
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.manifold import TSNE

#Cargamos los datos
iris = datasets.load_iris()
X = iris.data
y = iris.target 
#Aplicamos los datos
tsne = TSNE(n_components=2, max_iter=1000, random_state=123)
iris_T = tsne.fit_transform(X)  
#Representamos los datos transformados
plt.scatter(iris_T[y==0, 0], iris_T[y==0, 1], marker="o")   
plt.scatter(iris_T[y==1, 0], iris_T[y==1, 1], marker="*")
plt.scatter(iris_T[y==2, 0], iris_T[y==2, 1], marker="s")
plt.xlabel("Componente 1")
plt.ylabel("Componente 2")  
plt.title("Datos transformados con t-SNE")
plt.show()
```
