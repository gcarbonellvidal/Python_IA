Reglas de asociación. Para este ejemplo, tenemos que instalar la libreria pycaret

```python
from pycaret.datasets import get_data
data = get_data("france")
```

```python
%pip install pycaret

from pycaret.arules import *
s = setup(data, transaction_id='InvoiceNo', item_id='Description',
          ignore_items=['POSTAGE'])
arules = create_model(metric="confidence", min_threshold=0.5, min_support=0.085)
arules.to_csv("arules.csv")    
```
