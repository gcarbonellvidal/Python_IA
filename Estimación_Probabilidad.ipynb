Algoritmos de estimaci√≥n de probabilidad

```python
#Cargamos los paquetes necesarios

from matplotlib import pyplot as plt 
import pandas as pd 
import seaborn as sns 

#Cargamos los datos
df = sns.load_dataset("geyser")

#Representamos los histogramas con diferente anchura

plt.subplot(2,2,1)
plt.hist(df.duration, bins=5)
plt.title("Histograma-5 bins")
plt.xlabel("Duracion")
plt.subplot(2,2,2)
plt.hist(df.duration, bins=10)
plt.title("Histograma-10 bins")
plt.xlabel("Duracion")
plt.subplot(2,2,3)
plt.hist(df.duration, bins=20)
plt.title("Histograma-20 bins")
plt.xlabel("Duracion")
plt.subplot(2,2,4)
plt.hist(df.duration, bins=100)
plt.title("Histograma-100 bins")
plt.xlabel("Duracion")
plt.show()
```

```python
#Planteamos la modelizaci√≥n por n√∫cleos
sns.kdeplot(df.duration)
plt.title("Densidad de probabilidad")
plt.xlabel("Duracion")
plt.show()
```

```python
sns.set_style("whitegrid")
sns.kdeplot(x=df.duration, y=df.waiting)
plt.xlabel("Duracion", fontsize=14) 
plt.ylabel("Tiempo entre erupciones", fontsize=14)
plt.title("Relaci√≥n entre duraci√≥n y tiempo de espera", fontsize=16)
plt.show()
```
