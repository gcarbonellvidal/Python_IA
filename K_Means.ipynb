PRUEBAS CON EL ALGORITMO K-MEANS

```python
#Paquetes a usar para generar/visualizar los datos
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

#Generar datos sintéticos
np.random.seed(111)  # Para reproducibilidad
X, y = make_blobs(n_samples=150, centers=4, cluster_std=0.5)

#Visualización del conjunto de datos
plt.figure(0)
plt.grid(True)
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
#Libreria para el K-means
from sklearn.cluster import KMeans  
grupos=KMeans(n_clusters=2)  # Crear el modelo K-means
grupos.fit(X)  # Ajustar el modelo a los datos
centros = grupos.cluster_centers_  # Obtener los centros de los clusters
labels = grupos.labels_  # Obtener las etiquetas de los clusters   
plt.scatter(X[labels == 0, 0], X[labels == 0, 1], marker='o')
plt.scatter(X[labels == 1, 0], X[labels == 1, 1], marker='s')
plt.scatter(centros[:, 0], centros[:, 1], marker='x', color='black')
plt.show()
```

Calculamos el índice Silhoutte para comprobar el número óptimo de grupos.

```python
#libreria para calcular el índice Silhoutte
from sklearn.metrics import silhouette_score
centros = range(2,10)
ind=[]

for numero in centros:
    modelo = KMeans(n_clusters=numero)
    labels = modelo.fit_predict(X)
    silhouette_medio = silhouette_score(X, labels)
    ind.append(silhouette_medio)
    
fig, ax = plt.subplots(1, 1, figsize=(6, 6))
ax.plot(centros, ind, marker='o')
ax.set_title('Silhouette Score vs Number of Clusters')
ax.set_xlabel('Number of Clusters') 
ax.set_ylabel('Silhouette Score')
plt.show()
```

En el gráfico podemos observar que el número idóneo de puntos es 4. Pero podemos jugar un poco más con esto. Por ejemplo, ¿Qué ocurre si es estandarizan los datos de entrada, cambian los datos?
