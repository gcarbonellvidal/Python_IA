Vamos a usar el agrupamiento jerárquico para algoritmos dendograma

```python
#Librerias que vamos a usar

import numpy as np 
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.datasets import make_blobs

#Generamos unos datos y los representamos

np.random.seed(111)
X,y = make_blobs(n_samples = 20, cluster_std=0.7, centers=4)
plt.figure(0)
plt.grid(True)
plt.scatter(X[:,0],X[:,1])
plt.show()
```

A continuación generamos el dendograma de estos datos usando el método de enlace único(single linkage)

```python
dendograma = sch.dendrogram(sch.linkage(X, method='single'))
plt.title('Dendograma con enlace único')
plt.xlabel('Grupos')
plt.ylabel('Distancia')
plt.show()
```

A continuación, determinamos los agrupamientos

```python
#Librerias
from sklearn.cluster import AgglomerativeClustering
#Creamos el modelo de agrupamiento jerárquico   
modelo = AgglomerativeClustering(n_clusters=4, linkage='single')
#Ajustamos el modelo a los datos    
modelo.fit(X)
plt.title('Agrupamiento jerárquico')
#Representamos los datos
labels = modelo.labels_
plt.scatter(X[labels == 0, 0], X[labels == 0, 1], s=50, marker='o', label='Grupo 1')
plt.scatter(X[labels == 1, 0], X[labels == 1, 1], s=50, marker='s', label='Grupo 2')
plt.scatter(X[labels == 2, 0], X[labels == 2, 1], s=50, marker='^', label='Grupo 3')
plt.scatter(X[labels == 3, 0], X[labels == 3, 1], s=50, marker='x', label='Grupo 4')
plt.legend()
```
